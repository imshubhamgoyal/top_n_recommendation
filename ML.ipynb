```
from google.colab import drive 
drive.mount('/content/drive/')

from scipy import spatial

import numpy as np 
import pandas as pd
from pandas.tools.plotting import scatter_matrix
from random import randint
from sklearn.cluster import KMeans
import numpy as np
from scipy import spatial
from operator import itemgetter
from scipy.cluster.vq import vq, kmeans, whiten
import matplotlib.pyplot as plt
d1=dict()
user=set()
item=set()
with open('/content/drive/My Drive/CF project/u.data') as f:
    for line in f:
        l1=line.split('\t')
        u=int(l1[0])
        i=int(l1[1])
        rat=int(l1[2])
        user.add(u)
        item.add(i)
        d1[(u,i)]=rat
tt=int(len(user)/5)+1

mat=np.zeros((len(user),len(item)))

for ele in d1:
    u=ele[0]-1
    i=ele[1]-1
    rat=d1[ele]
    mat[u,i]=rat

no_of_cluster=10
kmeans = KMeans(n_clusters=10, random_state=0).fit(mat)
d1=dict()
clusters=[[]]*no_of_cluster
for i in range(len(kmeans.labels_)):
  label=kmeans.labels_[i]
  clusters[label].append(i)
  d1[i]=label


print(kmeans.labels_)
n=len(user)
m=len(item)

S=np.random.rand(m,m)
cluster=[]
for i in range(10):
  pg=np.random.rand(m,m)
  cluster.append(pg)
  
#learning rate
alpha=0.001
#regularization constants
gu=np.full((n,1),0.5)
g1u=np.full((n,1),0.5)

beta1=0.5
beta2=0.5

iter=100

global1=np.zeros((m,m))
print('here1')
for i in range(m):
  for j in range(m):
    if i==j:
      global1[i][j]=1
    else:
      l1=mat[:,i]
      l2=mat[:,j]
      s=1-spatial.distance.cosine(l1,l2)
      global1[i][j]=s
      
no_of_clusters=10
cl=[]
for ele in clusters:
  mat1=np.zeros((len(user),len(item)))
  for ele1 in ele:
    mat1[ele1,:]=mat[ele1,:]
  local=np.zeros((m,m))
  for i in range(m):
   for j in range(m):
    if i==j:
      local[i][j]=1
    else:
      l1=mat[:,i]
      l2=mat[:,j]
      s=1-spatial.distance.cosine(l1,l2)
      local[i][j]=s
  cl.append(local)

##leave one out cross validation 

top_n=20
test=randint(0, m)
count=0
count1=0
count2=0
count3=0
for i in range(n):
  pg=[]
  for j in range(m):
    rat=0
    for k in range(m):
      if mat[i][k]!=0:
        rat+=(gu[i]*global1[i][k])+(1-gu[i])*local[i][j]
    d2[j]=rat
    pg.append((j,rat))
  pg.sort(key=lambda x: x[1],reverse=True)
  a=pg.index((test,d2[test]))
  if a<10:
    count+=1
  if a<20:
    count1+=1
  if a<30:
    count2+=1
print("HR ",count/n)


#for i in range(no_of_cluster):

#whitened = whiten(mat)

#codebook, distortion = kmeans(mat, 5)
#print(codebook.shape)
#print(whitened.shape)

#codebook=codebook.transpose()
#data = pd.DataFrame(codebook, columns=['x1', 'x2', 'x3','x4','x5'])

#print(codebook[0])
#,'x6','x7','x8','x9','x10'])

#now plot using pandas 
#scatter_matrix(data, alpha=0.2, figsize=(6, 6), diagonal='kde')

#plt.scatter(codebook[:, 0], codebook[:, 1])
#plt.scatter(codebook[:, 2], codebook[:, 3])
#plt.show()

```
